## Aprendizado de maquina

```python
# Importando as bibliotecas para o aprenzado de mquina
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
```

Importando as bibliotecas para fazer o uso de tcnicas de aprendizagem, separao dos dados para teste e treino para fazer a anlise do modelo atravs da acurcia e preciso.




```python
train, test = train_test_split(dados, test_size = 0.3)

print(train.shape)
print(test.shape)
```

Dividindo  dados para teste e treino.

```python
# Dividindo dados de teste e treino
train_X = train[['Tamanho_do_Bico_mm','Profundidade_do_Bico_mm','Tamanho_da_Nadadeira_mm','Massa_Corporal_g']]
train_Y = train.especie

test_x = test[['Tamanho_do_Bico_mm','Profundidade_do_Bico_mm','Tamanho_da_Nadadeira_mm','Massa_Corporal_g']]
test_y = test.especie
```

Definindo as colunas para treino e teste

### Aprendizado com SVM

```python
# Definindo modelo para o aprendizado com SVM
modelo = svm.SVC()
modelo.fit(train_X,train_Y)
```

Neste bloco faz-se o treinamento de dados.

#### Fazendo uso da predio para gerar rtulos

```python
# Fazendo predio para gerar rtulos
predicao = modelo.predict(test_x)
```

Em seguida, calcular a acurcia do modelo

```python
# Calculando acurcia do modelo
acuracia= accuracy_score(y_true=test_y, y_pred=predicao)
print('A acurcia desse modelo :', acuracia * 100)
```

Observa-se  que a acurcia do modelo no  muito elevada e tambem, no  muito baixa.

```python
# Importando a biblioteca
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Exibindo  a preciso
print('A preciso do SVM :')
print(metrics.classification_report(test_y, predicao));
```

Aqui, a preciso para a epcie Adelie foi boa, o que indica que muitos dos dados analisados foram classificados da forma correta. A da Chinstrap no pode ser calculada, e da Gentoo, tem uma tima preciso em relao ao das outras espcies.

```python
# Usando a matriz de calor  com heatmap
plt.figure(figsize=(10,6))
sns.heatmap(dados.corr(),annot=True);
```

Os valores prximos de 1, seja negativo ou positivo, indicam que os atributos possuem correlao forte, portanto, quando os valores  de um atributo aumenta, o outro atributo tambm tende a aumentar quando a relatividade  positiva, e tende a diminuir quando sua relatividade  negativa.De mesma forma, notamos uma tendncia a aumentar da massa corporal e das caracteristicas de  tamanho do bico conforme o cescimento tendencial do tamanho da nadadeira.
```python
# Colunas a serem comparadas nas etapas de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

#### Analisando a acurcia e preciso

```python
# Gerando anlise de resultado de acurcia
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(X_train , y_train.values.ravel())
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes) * 100
print("A acurcia  de %.2f%% " % acuracia)
```

Na linha de modelo.fit(X_train, y_train.value.ravel()), pode-se analisar que foi adicionado o .value e o parmetro .ravel(). O.values vai fornecer esse valor num array(n,1) e o .ravel() converte esse array para (n,), fazendo com que aquele erro no acontea novamente.

```python
# Fazendo importao e uso da matriz de confuso , para identificar a correlao ,
# dos parmetros da espcie Chintrap, cujo os parmetros de "precision e recall" deram valores zerados,
# valores relativamente de alta diferena , em comparao  as outras espcies.

from sklearn.ensemble import RandomForestClassifier 

floresta_aleatoria = RandomForestClassifier( n_estimators=233 )
floresta_aleatoria.fit( X_train, y_train )

previsoes = floresta_aleatoria.predict( X_test)
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(y_test, previsoes)
print(matriz_confusao)

plt.figure(figsize=(10,5))
sns.heatmap(matriz_confusao, annot=True);
```

* Leitura e interpretao da tabela de confuso:

  * No eixo "X", faz-se a leitura dos valores previstos, atraves do aprendizado de mquina.
  * No eixo "Y", faz-se a leitura dos valores atuais, apresentados para a averiguao da acertividade da tomada de deciso apresentatos pelo modelo de comparao.

  * Os valores da tabelas ,denominados de: 0, 1 e 2 ,so correspondentes as trs diferentes espcies de pinguins; Adlie ,Chinstrap e Gentoo, respectivamente.

Conforme o que foi apresentatado na tabela, identifica-se que h boa acertividade entre o cruzamento dos respectivos valores de suas colunas e campos horizontais, denotando ser baixo o seu ndice de confuso em classificar os indivduos em suas respectivas especies.

```python
# Fazendo uso da rvore de deciso para maior acertividade da accurcia
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes) * 100
print("A preciso  de %.2f%%" % acuracia)
```

Fazendo o uso da vore de deciso, com base nas dimenses coletadas atravs da tabela de confuso ,obtermos a porcentagem  de preciso com maior acertividade e parmetros mais concretos.
