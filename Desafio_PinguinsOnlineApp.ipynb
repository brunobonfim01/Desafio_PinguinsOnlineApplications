## Aprendizado de maquina

```python
# Importando as bibliotecas para o aprenzado de máquina
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics 
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
```

Importando as bibliotecas para fazer o uso de técnicas de aprendizagem, separação dos dados para teste e treino para fazer a análise do modelo através da acurácia e precisão.




```python
train, test = train_test_split(dados, test_size = 0.3)

print(train.shape)
print(test.shape)
```

Dividindo  dados para teste e treino.

```python
# Dividindo dados de teste e treino
train_X = train[['Tamanho_do_Bico_mm','Profundidade_do_Bico_mm','Tamanho_da_Nadadeira_mm','Massa_Corporal_g']]
train_Y = train.especie

test_x = test[['Tamanho_do_Bico_mm','Profundidade_do_Bico_mm','Tamanho_da_Nadadeira_mm','Massa_Corporal_g']]
test_y = test.especie
```

Definindo as colunas para treino e teste

### Aprendizado com SVM

```python
# Definindo modelo para o aprendizado com SVM
modelo = svm.SVC()
modelo.fit(train_X,train_Y)
```

Neste bloco faz-se o treinamento de dados.

#### Fazendo uso da predição para gerar rótulos

```python
# Fazendo predição para gerar rótulos
predicao = modelo.predict(test_x)
```

Em seguida, calcular a acurácia do modelo

```python
# Calculando acurácia do modelo
acuracia= accuracy_score(y_true=test_y, y_pred=predicao)
print('A acurácia desse modelo é:', acuracia * 100)
```

Observa-se  que a acurácia do modelo não é muito elevada e tambem, não é muito baixa.

```python
# Importando a biblioteca
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Exibindo  a precisão
print('A precisão do SVM é:')
print(metrics.classification_report(test_y, predicao));
```

Aqui, a precisão para a epécie Adelie foi boa, o que indica que muitos dos dados analisados foram classificados da forma correta. A da Chinstrap não pode ser calculada, e da Gentoo, tem uma ótima precisão em relação ao das outras espécies.

Os valores próximos de 1, seja negativo ou positivo, indicam que os atributos possuem correlação forte, portanto, quando os valores  de um atributo aumenta, o outro atributo também tende a aumentar quando a relatividade é positiva, e tende a diminuir quando sua relatividade é negativa.De mesma forma, notamos uma tendência a aumentar da massa corporal e das caracteristicas de  tamanho do bico conforme o cescimento tendencial do tamanho da nadadeira.
```python
# Colunas a serem comparadas nas etapas de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

#### Analisando a acurácia e precisão

```python
# Gerando análise de resultado de acurácia
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(X_train , y_train.values.ravel())
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes) * 100
print("A acurácia é de %.2f%% " % acuracia)
```

Na linha de modelo.fit(X_train, y_train.value.ravel()), pode-se analisar que foi adicionado o .value e o parâmetro .ravel(). O.values vai fornecer esse valor num array(n,1) e o .ravel() converte esse array para (n,), fazendo com que aquele erro não aconteça novamente.

```python
# Fazendo importação e uso da matriz de confusão , para identificar a correlação ,
# dos parâmetros da espécie Chintrap, cujo os parâmetros de "precision e recall" deram valores zerados,
# valores relativamente de alta diferença , em comparação  as outras espécies.

from sklearn.ensemble import RandomForestClassifier 

floresta_aleatoria = RandomForestClassifier( n_estimators=233 )
floresta_aleatoria.fit( X_train, y_train )

previsoes = floresta_aleatoria.predict( X_test)
from sklearn.metrics import confusion_matrix

matriz_confusao = confusion_matrix(y_test, previsoes)
print(matriz_confusao)

plt.figure(figsize=(10,5))
sns.heatmap(matriz_confusao, annot=True);
```

* Leitura e interpretação da tabela de confusão:

  * No eixo "X", faz-se a leitura dos valores previstos, atraves do aprendizado de máquina.
  * No eixo "Y", faz-se a leitura dos valores atuais, apresentados para a averiguação da acertividade da tomada de decisão apresentatos pelo modelo de comparação.

  * Os valores da tabelas ,denominados de: 0, 1 e 2 ,são correspondentes as três diferentes espécies de pinguins; Adélie ,Chinstrap e Gentoo, respectivamente.

Conforme o que foi apresentatado na tabela, identifica-se que há boa acertividade entre o cruzamento dos respectivos valores de suas colunas e campos horizontais, denotando ser baixo o seu índice de confusão em classificar os indivíduos em suas respectivas espeécies.

```python
# Fazendo uso da árvore de decisão para maior acertividade da accurácia
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
previsoes = modelo.predict(X_test)

acuracia = accuracy_score(y_test, previsoes) * 100
print("A precisão é de %.2f%%" % acuracia)
```

Fazendo o uso da ávore de decisão, com base nas dimensões coletadas através da tabela de confusão ,obtermos a porcentagem  de precisão com maior acertividade e parâmetros mais concretos.
